1. Faça um programa que leia um número e, caso ele seja maior ou igual a zero, calcule e
mostre o número digitado ao quadrado. Caso ele seja menor que zero, mostre o dobro deste
número.

```python
num = float(input("Digite um número: "))

if num >= 0:
  print(f'O quadrado do número é {num**2}')
else:
  print(f'O dobro do número é {num*2}')
```


2. Faça um programa que peça um número inteiro e determine se ele é par ou ímpar. Dica:
utilize o operador módulo (resto da divisão).


```python
num = int(input('Digite um número: '))
if num%2 == 0:
  print(f'{num} é par')
else:
  print(f'{num} é ímpar')
```

3. Escreva um programa que leia 3 números e imprima o maior e o menor na mesma
mensagem.

```python
num1 = int(input('Digite um número: '))
num2 = int(input('Digite mais um número: '))
num3 = int(input('Digite outro número: '))

maior = num1
menor = num1

if num2 > num3 and num2 > num1:
  maior = num2
elif num3 > num2 and num3 > 1:
  maior = num3

if num2 < num1 and num2 < num3:
  menor = num2
elif num3 < num2 and num3 < num1:
  menor = num3

print(f'O menor número é {menor} e o maior número é {maior}')
```

4. Faça um programa para a leitura de duas notas parciais de um aluno. O programa deve
calcular a média alcançada por aluno e apresentar:
a) A mensagem "Aprovado", se a média alcançada for maior ou igual a sete;
b) A mensagem "Reprovado", se a média for menor do que sete;
c) A mensagem "Aprovado com Distinção", se a média for igual a dez.

```python
nota1 = float(input('Digite a primeira nota: '))
nota2 = float(input('Digite a segunda nota: '))

media = (nota1 + nota2)/2

if media >= 7 and media < 10:
  print('Aprovado!')
elif media == 10:
  print('Aprovado com distinção!')
else:
  print('Reprovado!')
```

5. Construa um programa para determinar se o indivíduo está com um peso favorável. Essa
situação é determinada através do IMC (Índice de Massa Corpórea), que é definida como
sendo a relação entre o peso (PESO) e o quadrado da Altura (ALTURA) do indivíduo
obedecendo a seguinte fórmula:

image.png

```python
peso = float(input('Digite o peso: '))
altura = float(input('Digite a altura: '))

imc = peso / (altura ** 2)
print(imc)

if imc >= 40:
  print ('Obeso mórbido')
elif 40 > imc >= 30:
  print('Obeso')
elif 30 > imc >= 25:
  print('Sobrepeso')
elif 25 > imc >= 20:
  print('Peso normal')
elif imc < 20:
  print('Abaixo do peso')

```
