apenas ela lembra abastecer a garrafa, causando aquele inconveniente diário. Luciene toma 2
xícaras de café por dia, sabendo que a cafeteira fornece 10 xícaras por dia. Faça um programa
que receba como entrada a quantidade de xícaras de café que as colegas de Luciene tomaram

2. Um deputado propôs um projeto para alterar as regras para a aposentadoria. Por este
projeto, para requerer a aposentadoria, os trabalhadores têm que combinar dois requisitos:
tempo de contribuição ao INSS e idade mínima. Os trabalhadores do sexo masculino poderão
aposentar-se com no mínimo 50 anos de idade e no mínimo 30 anos de contribuição. Além
disso, é necessário que a soma entre o tempo de contribuição e a idade seja de no mínimo 90
anos para eles.
Faça um programa que leia a idade e o tempo de contribuição de um trabalhador do sexo
masculino e informe se o mesmo pode se aposentar. Não é necessário validar a idade e o
tempo de contribuição.

```python
idade = int(input('Digite a idade: '))
tempo_de_contribuicao = int(input('Digite o tempo de contribuição: '))

if tempo_de_contribuicao >= 30 and idade >= 50:
  print('O trabalhador pode se aposentar')
else:
  print('O trabalhador não pode se aposentar')
```

3. A prefeitura de Ouro Preto contratou você para fazer um programa que calcule os valores
do IPTU de imóveis da cidade, conforme o tipo do loteamento e a área dos mesmos. Deverão
ser considerados apenas dois tipos de loteamento: 1 e 2. Para cada tipo de loteamento, se a
área do imóvel for menor que 200 m2, efetua-se um cálculo de IPTU; se for maior ou igual a
200 m2, efetua-se outro cálculo de IPTU. A tabela abaixo mostra como o cálculo deve ser
efetuado para cada caso.

image.png

Faça um programa que leia o tipo de um loteamento e a área do mesmo e apresente o valor do
IPTU de um determinado imóvel de Ouro Preto, calculado conforme a tabela acima.

```python
tipo_de_loteamento = int(input('Digite o tipo de loteamento: '))
area_do_loteamento = int(input('Digite a área do loteamento: '))

if tipo_de_loteamento == 1:
  if 0 < area_do_loteamento < 200:
    valor_do_iptu = area_do_loteamento * 1.0
  else:
    valor_do_iptu = area_do_loteamento * 1.2
else:
  if 0 < area_do_loteamento < 200:
    valor_do_iptu = area_do_loteamento * 1.1
  else:
    valor_do_iptu = area_do_loteamento * 1.3

print(f'O valor do iptu é {valor_do_iptu}')
```

4. Escreva um programa que leia a nota final de um aluno referente à disciplina de Introdução
à Programação. Caso a nota seja maior ou igual a 6.0, o programa imprime uma mensagem
dizendo que o aluno foi aprovado.
No caso da nota ser menor que 6.0, o programa imprime uma mensagem informando que o
aluno está em exame especial, e faz uma nova leitura de nota desse aluno, referente à nota do
exame especial. Caso a nota do exame especial seja maior ou igual a 6,0, o programa imprime a mensagem que o aluno foi aprovado; caso contrário, imprime que o aluno foi reprovado.

```python
nota = float(input('Digite a nota final do aluno: '))

if nota >= 6.0:
    print('O aluno foi aprovado')
else:
    print('O aluno está em exame especial')
    nota_exame = float(input('Digite a nota do exame especial: '))
    
    if nota_exame >= 6.0:
        print('O aluno foi aprovado')
    else:
        print('O aluno foi reprovado')
```

1. Faça um programa que recebe o salário de um colaborador e o reajuste segundo o
seguinte critério, baseado no salário atual:

  1) salários até R$ 280,00 (incluindo) : aumento de 20%

  2) salários entre R$ 280,00 e R$ 700,00 : aumento de 15%

  3) salários entre R$ 700,00 e R$ 1500,00 : aumento de 10%

  4) salários de R$ 1500,00 em diante : aumento de 5%



Após o aumento ser realizado, informe na tela:

  a) o salário antes do reajuste;

  b) o percentual de aumento aplicado;

  c) o valor do aumento; 
  
  d) o novo salário, após o aumento.

```python
salario = float(input('Digite o salário: '))
if salario <= 280:
  reajuste = 20
elif 280 < salario <= 700:
  reajuste = 15
elif 700 < salario <= 1500:
  reajuste = 10
else:
  reajuste = 5
  
aumento = salario * reajuste / 100

novo_salario = salario + aumento

print(f'Salário antes do reajuste: {salario}. Percentual de aumento: {reajuste}%. Valor do aumento: {aumento}. Novo salário: {novo_salario}')
```

3. Uma servidora que atende o público externo tem seu desempenho medido pela
quantidade de processos despachados ao longo do mês. Para que possa ser elegível ao
prêmio "servidora destaque", ela precisa despachar pelo menos 100 processos/mês. Faça
um programa que receba como entrada o número de processos que a servidora
despachou no dia, em seguida calcule quantos processos foram despachados no mês e se
ela é uma servidora elegível para o prêmio "servidora destaque". (Obs: atenção a
quantidade de dias úteis de um mês).

```python
num_processos_dia = int(input('Digite o número de processos que a servidora despachou no dia: '))

processos_mes = 22 * num_processos_dia

if processos_mes >= 100:
  print('Servidora elegível para o prêmio "servidora destaque"')
```

4. A servidora Maria tem 15 minutos de intervalo por dia. Em cada intervalo ela assiste
15 minutos do seu drama favorito "A advogada extraordinária". Sabendo que cada
episódio tem 45 minutos e o drama possui 14 episódios, faça um programa que receba a
quantidade de dias trabalhados, retorne quantos episódios foram finalizados e se ela
precisa de uma nova série para assistir.

```python
dias_trabalhados = int(input('Digite a quantidade de dias trabalhados: '))

episodios = dias_trabalhados * 15 / 45

print(f'Episódios finalizados {episodios}')

if episodios >= 14:
  print('Ela precisa de uma nova série')
```
